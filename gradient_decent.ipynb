### Gradient Descent in Linear Regression

#### Gradient descent is an optimization technique applied in linear regression to reduce prediction errors. Let me explain how gradient descent functions within the context of linear regression.

#### Why Use Gradient Descent in Linear Regression?
Linear regression aims to find the optimal line that best fits a dataset by minimizing the error between the actual and predicted values. This error is quantified using a cost function, commonly the Mean Squared Error (MSE). The objective is to determine the model parameters (slope, ùëö and intercept b) that minimize this cost function.

In the case of simple linear regression, we can directly calculate the parameters using formulas like the Normal Equation (y = mx + b). However, for large datasets or high-dimensional data, these methods become computationally expensive or even impractical.

## How Gradient Descent Works in Linear Regression:

1. **Initialize Parameters**: Begin by assigning random initial values for the slope \(m\) and the intercept \(b\).

2. **Compute the Cost Function**: Calculate the error between the predicted values and the actual values using a cost function like Mean Squared Error (MSE):
3. **Calculate the Gradient**: Determine the gradient of the cost function with respect to \(m\) (slope) and \(b\) (intercept). These gradients reveal how the cost changes when the parameters are adjusted:
4. **Update the Parameters**: Adjust \(m\) and \(b\) in the direction that reduces the cost, based on the gradients:
5. **Iterate**: Repeat the process until the cost function converges, meaning further updates yield negligible changes in the cost.


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression

# Generate synthetic regression dataset
X, y = make_regression(n_samples=200, n_features=1, noise=10, random_state=0)
y = y.reshape(-1, 1)
m = X.shape[0]

X_b = np.c_[np.ones((m, 1)), X]

theta = np.array([[2.0], [3.0]])

plt.figure(figsize=(6, 8))
plt.scatter(X, y, color="blue", label="Actual Data")
plt.plot(X, X_b.dot(theta), color="red", label="Initial Line (No GD)")
plt.xlabel("Feature")
plt.ylabel("Target")
plt.title("Linear Regression Without Gradient Descent")
plt.legend()
plt.show()
```

### Here we can see that model is not making accurate predictions, and the line is not fitting the data properly. This is due to suboptimal parameter initialization. It prevents the algorithm from converging to the best-fit line.

### Lets use gradient descent in this model to optimize it.

```python
learning_rate = 0.1
n_iterations = 150

# Performing Gradient Descent
for _ in range(n_iterations):
    # Predictions
    y_pred = X_b.dot(theta)
    # Computing the gradient
    gradients = (2 / m) * X_b.T.dot(y_pred - y)
    # Updating parameters
    theta -= learning_rate * gradients

plt.figure(figsize=(10, 5))
plt.scatter(X, y, color="blue", label="Actual Data")
plt.plot(X, X_b.dot(theta), color="red", label="Optimized Line (With GD)")
plt.xlabel("Feature")
plt.ylabel("Target")
plt.title("Linear Regression With Gradient Descent")
plt.legend()
plt.show()
```

##### Linear Regression using Gradient Descent demonstrates how the model gradually learns to fit the line that minimizes the error between predicted and actual values by updating the parameters iteratively.
##### This method is a straightforward optimization technique that fine-tunes model parameters to identify the best-fit line for a given dataset. By continuously minimizing the cost function, gradient descent ensures that the line closely represents the relationship between the variables. With appropriate parameter tuning, gradient descent provides reliable and precise predictions across various applications.
